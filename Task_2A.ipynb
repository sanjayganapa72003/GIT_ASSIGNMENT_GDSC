```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("nlp_intel_test.csv",header=None)
```

```python
data.columns = ["num","english","french"]
```

```python
data = data.iloc[1:,1:2]
```

```python
#we have to tokenize the sentences and remove all the punctuations present in a sentences
import re
import string
def Tokenize_and_rem_punc(text):
  text_no_punc="".join([char for char in text if char not in string.punctuation])
  text = re.split("\W",text_no_punc)
  return text
data['text_clean']=data['english'].apply(lambda x:Tokenize_and_rem_punc(x.lower()))
```

```python
# now we can calculate frequency of words appeared in a sentence
def frequency(text):
  count_list=[]
  for i in text:
    count_list.append(text.count(i))
  return count_list

data["frequency"] = data["text_clean"].apply(lambda x:frequency(x))
```

```python
# now we calculate the length of the words contributing to the length of the sentence
def length(text):
  length_list=[]
  for i in text:
    length_list.append(len(i)/len("".join(text)))
  return length_list
data["length"] = data["text_clean"].apply(lambda x:length(x))
```

```python
#And now we finally multiply its length and frequnency and get the final word embedding
word_embedding = lambda list1,list2 : [x*y for x,y in zip(list1,list2)]
data["word_embedding"]= data.apply(lambda x: word_embedding(x["frequency"],x["length"]),axis=1)
```

```python
# Dataframe containing english sentences and its corresponding word embeddings
Embedded_data = data.drop(["frequency","text_clean","length"],axis=1)
Embedded_data
```
